## ðŸŽ² Estimating Pi with Monte Carlo
Weâ€™ll now estimate Pi using random sampling.

```python
import random

def estimate_pi(num_samples):
    inside_circle = 0
    for _ in range(num_samples):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        if x**2 + y**2 <= 1:
            inside_circle += 1
    return (4 * inside_circle) / num_samples

# Estimate Pi using 10000 points
print("Estimated Pi:", estimate_pi(10000))
```

### Description of the Hailstone sequence model

This code generates the **Hailstone sequence** starting from `N = 71`. The Hailstone sequence is defined as:

- If `N` is even: divide `N` by 2.
- If `N` is odd: compute `3N + 1`.

This process repeats until `N` becomes 1.

- **Output:**
  - The code prints each term of the sequence.
  - It also counts and prints the number of steps required to reach 1.

This sequence is called *"hailstone sequence"* because the values rise and fall like hailstones in a cloud before eventually settling.  
The sequence is associated with the famous **Collatz conjecture**, which asserts that this process will reach 1 for *any positive integer starting value*, though this has not been proven for all integers.


```python
N = 17
steps=0
while N != 1:
    print(N)
    steps += 1
    if N%2 == 0:
        N=N//2
    else:
        N = 3*N+1
print("steps=",steps)
```

### SIR model equations

The classic **SIR model** is defined by the following system of differential equations:

$$
\begin{aligned}
\frac{dS}{dt} &= -\alpha S I \\
\frac{dI}{dt} &= \alpha S I - \gamma I \\
\frac{dR}{dt} &= \gamma I
\end{aligned}
$$

where:
- \( S(t) \) is the number of susceptible individuals,
- \( I(t) \) is the number of infected individuals,
- \( R(t) \) is the number of recovered individuals,
- \( \alpha \) is the transmission rate (per person per unit time),
- \( \gamma \) is the recovery rate (per unit time).


### Complicated dynamics
# SIR
### Description of the SIR model implemented

This code implements a simple **SIR (Susceptible-Infected-Recovered) model** for simulating the spread of an infectious disease in a closed population over 20 days.

- **Model parameters:**
  - `alpha`: transmission rate (per person per day)
  - `gamma`: recovery rate (per day)
  - Initial conditions: 
    - Susceptible (`S = 762`)
    - Infected (`I = 1`)
    - Recovered (`R = 0`)

- **Dynamics:**
  - At each time step (1 day), the number of individuals in each compartment is updated using:
    - Î”S = -Î± S I
    - Î”I = Î± S I - Î³ I
    - Î”R = Î³ I

- **Visualization:**
  - The history of S, I, and R is plotted over time:
    - Susceptible (blue)
    - Infected (red)
    - Recovered (green)

This is a **discrete-time approximation** of the classic continuous SIR model and provides insight into the epidemic dynamics.


```python
import matplotlib.pyplot as plt
alpha=0.00218 #(person-1 day-1)
gamma=0.88    #(day-1)
S=762
I=1
R=0
S_hist=[]
I_hist=[]
R_hist=[]
for day in range(20):
    S_hist.append(S)
    I_hist.append(I)
    R_hist.append(R)
    delta_S=(-alpha*S*I)
    delta_I=alpha*S*I-gamma*I
    delta_R=gamma*I
    S+=delta_S
    I+=delta_I
    R+=delta_R
    S=max(S,0)
    I=max(I,0)
    R=max(R,0)
#Plot S,I,R vs day
plt.plot(S_hist)
plt.plot(I_hist,color='red')
plt.plot(R_hist,color='green')
plt.show()
```

## Morgage payment
### Description of the mortgage payment model

This code simulates the payoff of a **mortgage loan** with fixed monthly payments and interest over time.

- **Model parameters:**

  - Monthly payment: \$600

- **Dynamics:**

```python
import matplotlib.pyplot as plt
intrst=0.06/12
payment=600
bal=[60000.] #Initial balance
for month in range(1000):
    bal.append(bal[-1]*(1+intrst)-payment)
    if bal[-1]<=0: break
plt.plot(bal)
title="Initial Balance=%8.2f Total paid=%8.2f Months=%3.1f Years=%2.2f"%(bal[0],month*payment,month, month/12)
plt.xlabel("Months")
plt.ylabel("Balance")
plt.title(title)
plt.show()
```

## Saving
### Description of the savings accumulation model

This code simulates the accumulation of savings in a bank account with **regular monthly deposits** and **compound interest**, over a period of 40 years (480 months).

- **Model parameters:**
  - Monthly interest rate: 0.098 / 12 â‰ˆ 0.00817% per month (approximately 9.8% annual rate)
  - Monthly deposit: \$50
  - Duration: 480 months (40 years)

- **Dynamics:**
  - For each month:
    - The balance grows by accruing interest on the current balance.
    - A fixed deposit (`payment`) is added.
  - The result is **compound growth** of the balance over time.

- **Visualization:**
  - The plot shows the balance versus time:
    - Primary x-axis: months.
    - Secondary x-axis: years (labeled every 5 years).
  - A grid improves readability, and the title displays the initial and final balances.

This model illustrates how **regular contributions and compound interest together lead to substantial growth over the long term**, a key principle in personal finance.


```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters
interest = 0.098 / 12  # Monthly interest rate
payment = 50
months = 480  # 40 years

# Balance calculation
bal = [0.0]
for month in range(months):
    new_balance = bal[-1] * (1 + interest) + payment
    bal.append(new_balance)

# Create x-axis values
month_vals = np.arange(0, months + 1)
year_vals = month_vals / 12

print(payment,interest*100, months/12)
# Plotting
fig, ax1 = plt.subplots(figsize=(6, 4))
ax1.plot(month_vals, bal, color='darkorange', linewidth=2)
ax1.set_xlabel("Month", fontsize=12)
ax1.set_ylabel("Balance ($)", fontsize=12)

# Title with initial and final balance
title = "Initial Balance = \${:,.2f}, Final Balance = \${:,.2f}".format(bal[0], bal[-1])
plt.title(title, fontsize=14)

# Add secondary x-axis for years
ax2 = ax1.twiny()
ax2.set_xlim(ax1.get_xlim())
ax2.set_xticks(np.arange(0, months + 1, 60))  # every 5 years
ax2.set_xticklabels([f"{int(x)} yr" for x in np.arange(0, 41, 5)])
ax2.set_xlabel("Years", fontsize=12)

# Grid and layout
ax1.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### Population dynamics
### Description of the exponential population growth model

This code simulates **exponential population growth** over 150 time steps.

- **Model parameters:**
  - Initial population: `pop = 100`
  - Growth rate: `r = 0.05` (per time step)

- **Dynamics:**
  - At each step, the population increases by a constant proportion:
    - Î”pop = r Ã— pop
  - This leads to **unbounded exponential growth**, reflecting a scenario with unlimited resources and no environmental constraints.

- **Visualization:**
  - The population over time is plotted as a continuous curve showing exponential increase.

This simple model is useful to illustrate population growth in idealized conditions, before introducing limiting factors (e.g., logistic growth with carrying capacity).


```python
import matplotlib.pyplot as plt
pop=100
pops=[]
r=0.05
for i in range(150):
	pops.append(pop)
	deltap=r*pop
	pop+=deltap
plt.plot(pops)
plt.ylabel("Population")
plt.xlabel("Time")
plt.show()
```
